#########Data cleaning

```python
vac_df.fillna(value = 0, inplace = True)
vac_df.total_vaccinations = vac_df.total_vaccinations.astype(int)
vac_df.people_vaccinated = vac_df.people_vaccinated.astype(int)
vac_df.people_fully_vaccinated = vac_df.people_fully_vaccinated.astype(int)
vac_df.daily_vaccinations_raw = vac_df.daily_vaccinations_raw.astype(int)
vac_df.daily_vaccinations = vac_df.daily_vaccinations.astype(int)
vac_df.total_vaccinations_per_hundred = vac_df.total_vaccinations_per_hundred.astype(int)
vac_df.people_fully_vaccinated_per_hundred = vac_df.people_fully_vaccinated_per_hundred.astype(int)
vac_df.daily_vaccinations_per_million = vac_df.daily_vaccinations_per_million.astype(int)
vac_df.people_vaccinated_per_hundred = vac_df.people_vaccinated_per_hundred.astype(int)
date = vac_df.date.str.split('-', expand =True)
date
```

```python
vac_df['year'] = date[0]
vac_df['month'] = date[1]
vac_df['day'] = date[2]
vac_df.year = pd.to_numeric(vac_df.year)
vac_df.month = pd.to_numeric(vac_df.month)
vac_df.day = pd.to_numeric(vac_df.day)
vac_df.date = pd.to_datetime(vac_df.date)
vac_df.head()

```

```python
vac_df.country.unique()
```

```python
country_wise_total_vaccinated = {}
for country in vac_df.country.unique() : 
    vaccinated = 0
    for i in range(len(vac_df)) : 
        if vac_df.country[i] == country : 
            vaccinated += vac_df.daily_vaccinations[i]
            country_wise_total_vaccinated[country] = vaccinated 
#     made a seperate dict from the df 
    country_wise_total_vaccinated_df = pd.DataFrame.from_dict(country_wise_total_vaccinated,
                                                         orient='index',
                                                         columns = ['total_vaccinted_till_date'])
#     converted dict to df 
country_wise_total_vaccinated_df.sort_values(by = 'total_vaccinted_till_date', ascending = False, inplace = True)
country_wise_total_vaccinated_df

```

```python
fig = px.bar(country_wise_total_vaccinated_df, 
             y = 'total_vaccinted_till_date',
             x = country_wise_total_vaccinated_df.index,
             color = 'total_vaccinted_till_date',
             color_discrete_sequence= px.colors.sequential.Viridis_r
            )
fig.update_layout(
    title={
            'text' : "Vaccination till date in various countries",
            'y':0.95,
            'x':0.5
        },
    xaxis_title="Countries",
    yaxis_title="Total vaccinated",
    legend_title="Total vaccinated"
)
fig.show()
```
